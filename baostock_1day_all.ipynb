# 用Basostock获取一天全部股票的信息

```python
# sd 开始日
# ed 结束日（主要日）
sd = '2022-07-06'
ed = '2022-07-07'
```

```python
import baostock as bs
import pandas as pd
#### 登陆系统 ####
lg = bs.login()
# 显示登陆返回信息
print('login respond error_code:'+lg.error_code)
print('login respond error_msg:'+lg.error_msg)
```

```python
class test:

    def __init__(self):
        self.data_list=[]
        self.rs=None

    def get(self,gpd):
        rs = bs.query_history_k_data_plus(
            gpd[0],"date,code,open,high,low,close,preclose,volume,amount,adjustflag,turn,tradestatus,pctChg,isST",
            start_date = sd, end_date = ed, frequency="d", adjustflag="3")
        while (rs.error_code == '0') & rs.next():
            self.data_list.append(rs.get_row_data())
            print(rs.get_row_data())
            self.rs=rs
```

```python
import time
lg = bs.login()
rs = bs.query_all_stock(day=ed)
te = test()
print("start")
i = 0
s_time = time.time()
while (rs.error_code == '0') & rs.next():
    i = i + 1
    if ((i % 100) == 0):
        print(i)
    gpd=rs.get_row_data()
    te.get(gpd)
result = pd.DataFrame(te.data_list, columns=te.rs.fields)
result.to_csv(r"D:\gitmaker\Fin_data\all.csv", index=False)
bs.logout()
c_time = time.time()
ss_time = c_time - s_time
print('%.3f秒'%ss_time)
```
